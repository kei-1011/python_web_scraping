```python
from newspaper import Article
# Article で単一の記事を取得
```

```python
URL = "https://jp.reuters.com/article/us-econ-payroll-idJPKBN28E26S"
# 記事URL
```

```python
article = Article(URL)
# article を代入
```

```python
article.download()
# ダウンロード
```

```python
article.parse()
# パース
```

```python
article.title
# 記事タイトル
```

```python
article.text
# 記事テキスト
```
