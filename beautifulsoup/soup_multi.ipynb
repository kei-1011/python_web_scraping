## Yahooの主要ニュースを取得

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import time
```

```python
url = 'https://www.yahoo.co.jp/'
```

```python
selector = '#tabpanelTopics1 > div > div._2jjSS8r_I9Zd6O9NFJtDN- > ul li a'
```

```python
res = requests.get(url)
```

```python
soup = BeautifulSoup(res.text,'html.parser')
```

```python
# 全てのaタグの情報を取得
elms = soup.find_all('a')
```

```python
# reモジュールを使用
import re
```

```python
# news.yahoo.co.jp/pickupが含まれているものに絞り込み
elms = soup.find_all(href=re.compile('news.yahoo.co.jp/pickup'))
```

```python
print(elms[0].span.string)
```

```python
print(elms[0].attrs['href'])
```

```python
# リストから取り出して表示
for elm in elms:
    print(elm.span.string)
    print(elm.attrs['href'])    
```

## 階層化されたリンクを辿って、記事を取得する

```python
# URLをリストに格納
pickup_links = [elm.attrs['href'] for elm in elms]
```

```python
print(pickup_links)
```

```python
# 要約ページのURLを１軒ずつ処理
for pickup_link in pickup_links:
    pickup_res = requests.get(pickup_link)
    pickup_soup = BeautifulSoup(pickup_res.text,'html.parser')
    pickup_elem = pickup_soup.find('p', class_='sc-OqFzE')
    news_link = pickup_elem.a.attrs['href']
    news_res = requests.get(news_link)
    news_soup = BeautifulSoup(news_res.text,'html.parser')
    print(news_soup.title.text)
    print(news_link)
    detail_text = news_soup.find(class_=re.compile('sc-gqPbQI'))

    # テキストがなければ、''を表示
    print(detail_text.text if hasattr(detail_text, 'text') else '',end='\n\n')
```
